```python
#Import necessary packages
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Open dataset and extract variables
datafile = 'data/gmsl.nc'
with xr.open_dataset(datafile) as ds:
    gmsl = ds['gmsl'].values
    time = ds['time'].values
    scens = ds['scen'].values
    model = ds['model'].values
```

```python
#Check the content of the variables
print(model)
print(scens)
print(gmsl.shape)
```

```python
#Make a quick plot
fig,ax = plt.subplots(1,2,sharey=True)
for s,scen in enumerate(scens):
    for m,mod in enumerate(model):
        ax[s].plot(time,gmsl[m,s,:])
    ax[s].set_title(scen)
```
