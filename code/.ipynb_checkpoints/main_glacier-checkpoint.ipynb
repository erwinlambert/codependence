```python
import numpy as np
import functions as fun
```

```python
def glacier(model,scen):
    #Read data
    with open(f'../rawdata/marzeion2012/dV_global_accum_{model}_{scen}_r1i1p1.txt','r') as file:
        for l,line in enumerate(file):
            y = int(float(line.split()[0]))
            if l == 0:
                y = int(float(line.split()[0]))
                years = np.array([y])
                glac = np.array([0])
            else:
                years = np.append(years,y)
                glac = np.append(glac,float(line.split()[1])/10.)

    #Detrend
    SLR = fun.detrend(glac,np.array([]),years)

    #Save SLR
    fun.saveSLR(SLR,years,model,scen,'glacier')
```

```python
def glacierregion(model,scen):
    for r in np.arange(1,20):
        with open(f'../rawdata/marzeion2012/dV_regional_accum_{model}_{scen}_r1i1p1.txt','r') as file:
            for l,line in enumerate(file):
                y = int(float(line.split()[0]))
                if l == 0:
                    y = int(float(line.split()[0]))
                    years = np.array([y])
                    glac = np.array([0])
                else:
                    years = np.append(years,y)
                    glac = np.append(glac,float(line.split()[r])/10.)

        #Detrend
        SLR = fun.detrend(glac,np.array([]),years)

        #Save SLR
        fun.saveSLR(SLR,years,model,f'{scen}_{r}','glacier')
```

```python
mods = fun.models()
for model in mods:
    for scen in ['rcp45','rcp85']:
        glacier(model,scen)
        glacierregion(model,scen)
```
